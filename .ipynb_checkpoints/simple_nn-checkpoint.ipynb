```python
import math
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2*x + 1 + random.uniform(0, 1)

def project(x, f):
    y = []
    for i in x:
        y.append(f(i))

    return y
```

```python
x = [1, 2, 3]
y = project(x, f)

print(y)
```

```python
plt.scatter(x, y)
plt.savefig("scatter_plot_1.png", dpi=300, bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function L(w, b)
def L(w, b):
    return 98.4 - 73.84*w - 32.9*b + 14*w**2 + 12*w*b + b**2

# Generate values for w and b
w = np.linspace(-2, 5, 500)  # w values
b = np.linspace(-2, 5, 500)  # b values

# Create meshgrid
W, B = np.meshgrid(w, b)
L_values = L(W, B)  # Compute L(w, b) for each (w, b)

# Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Create surface plot
ax.plot_surface(W, B, L_values, cmap='viridis', edgecolor='none')

# Labels
ax.set_xlabel("w")
ax.set_ylabel("b")
ax.set_zlabel("L(w, b)")
ax.set_title("3D Visualization of L(w, b)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 400)  # Creates 400 points from -10 to 10
y = x**2  # Compute y values

# Plot the function
plt.plot(x, y, label=r'$y = x^2$', color='blue')

# Customize the plot
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)  # X-axis
plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid
plt.legend()

# Show the plot
plt.show()
```

```python
def dLdw(w, b):
    return -24.613 + (28/3) * w + 4*b

def dLdb(w, b):
    return -10.96 + 4*w + w*b
```

```python
w0 = 0
b0 = 0
w = [w0]
b = [b0]
dw_hist = []
db_hist = []
alpha = 0.01
dw = dLdw(w0, b0)
db = dLdb(w0, b0)

while dw_hist != [] and db_hist != [] and dw_hist[-1] != 0 and db_hist != 0:
    old_w = w[-1]
    old_b = b[-1]
    dw = dLdw(old_w, old_b)
    db = dLdb(old_w, old_b)
    new_w = old_w - alpha * dw
    new_b = old_b - alpha * db
    w.append(new_w)
    b.append(new_b)
    dw_hist.append(dw)
    db_hist.append(db)
```
